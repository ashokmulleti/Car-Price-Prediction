```python
#Lets import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver

import time

#Importing requests
import requests

# importing regex
import re

# Importing required Exceptions
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException
from selenium.webdriver.support.ui import WebDriverWait

import warnings
warnings.filterwarnings('ignore')
```

```python
#to open the new window
driver = webdriver.Chrome(r"C:\Users\Ashok\Downloads\chromedriver_win32 (1)\chromedriver")
```

```python
driver.get('https://www.cars24.com/buy-used-car?sort=P&storeCityId=2&pinId=110001')
```

```python
SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")-100

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

### Scraping Car Name and Year

```python
Car_Name=[]
Year=[]

Model=driver.find_elements_by_xpath("//div[@class='_1l4fi']/div/div/h2")
for i in Model:
    y=i.text
    Car_Name.append(y[5:])
    Year.append(y[0:4])
        
print(Car_Name)
print('\n')
print(Year)
```

```python
len(Year)
```

```python
len(Car_Name)
```

### Scraping Car variant

```python
Car_variant=[]
try:
    c_var=driver.find_elements_by_xpath("//div[@class='_15rKe']/div/p")
    for i in c_var:
        Car_variant.append(i.text)
        
except NoSuchElementException:
    Car_variant.append("no detail available")
except StaleElementReferenceException:
    Car_variant.append("no detail available")
print(Car_variant)
```

```python
len(Car_variant)
```

### Scraping Current Price

```python
New_Price=[]
location=[]
try:
    n_price=driver.find_elements_by_xpath("//div[@class='_7udZZ']/span")
    for i in n_price:
        New_Price.append(i.text)
        location.append("New Delhi")
except NoSuchElementException:
    New_Price.append("no detail available")
except StaleElementReferenceException:
    New_Price.append("no detail available")
print(New_Price)
```
